```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#dataset
df = pd.read_csv('../data/advertising.csv')
df
```

```python
# Extracting the datetime variables using timestamp column

df['Timestamp'] = pd.to_datetime(df['Timestamp']) 
df['Month'] = df['Timestamp'].dt.month 
df['Day'] = df['Timestamp'].dt.day     
df['Hour'] = df['Timestamp'].dt.hour   
df["Weekday"] = df['Timestamp'].dt.dayofweek 

df['Date'] = df['Timestamp'].apply(lambda t : t.date())
df = df.drop(['Timestamp'], axis=1) # deleting timestamp
df
```

### Model building and testing on different ml algorithms 

```python
# Function to split data and train a model
def train_model(X, y, model, scaler=None, test_size=0.3, random_state=42):
    if scaler is not None:
        X = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    model.fit(X_train, y_train)
    return X_test, y_test, model

# Function to make predictions and generate classification report and confusion matrix
def evaluate_model(X_test, y_test, model):
    predictions = model.predict(X_test)
    print(classification_report(y_test, predictions))
    cm = confusion_matrix(y_test, predictions)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
    plt.show()


# Specify features and target
X = df[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male']]
y = df['Clicked on Ad']

# Creating and training Logistic Regression model
lgr = LogisticRegression()
X_test, y_test, lgr = train_model(X, y, lgr)
print("Logistic Regression:")
evaluate_model(X_test, y_test, lgr)

# Creating and training Decision Tree model
dt = DecisionTreeClassifier(random_state=42)
X_test, y_test, dt = train_model(X, y, dt)
print("Decision Tree:")
evaluate_model(X_test, y_test, dt)

# Creating and training Random Forest model
rf = RandomForestClassifier(random_state=42)
X_test, y_test, rf = train_model(X, y, rf)
print("Random Forest:")
evaluate_model(X_test, y_test, rf)
from sklearn.svm import SVC

# Creating and training Support Vector Machine (SVM) model
svm = SVC(kernel='linear', random_state=42)
X_test, y_test, svm = train_model(X, y, svm)
print("Support Vector Machine (SVM):")
evaluate_model(X_test, y_test, svm)
```
